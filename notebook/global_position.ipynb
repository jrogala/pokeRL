```python
from matplotlib import pyplot as plt
import numpy as np

from pokerl.env.pokemonblue import PokemonBlueEnv
```

```python
env = PokemonBlueEnv()
env.reset()
pos = set()
for _ in range(5000):
    action = env.action_space.sample()
    observation, reward, truncated, terminated, info = env.step(action)
    x,y = info["absolute_position"]
    pos.add((x,y))
    if truncated or terminated:
        env.reset()
env.close()
print(len(pos))
```

```python
pos_array = np.array(list(pos))
plt.scatter(pos_array[:,1], pos_array[:,0])
```
